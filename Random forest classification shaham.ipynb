```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#8*8 matrix -0 7
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i]) 
```

```python
digits.data[:5]
```

```python
df = pd.DataFrame(digits.data)#0-64 integers
df.head()
```

```python
df['target'] = digits.target
```

```python
df[0:12]
```

```python
X = df.drop('target',axis='columns')#x is a independent variable df alredy target so target drop cheynm
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)#20% train cheyipikkunu 80% test cheyipikkunnu
```

```python
len(X_train)
```

```python
len(X_test)#20% of total sample
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis='columns'),iris.target,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
model = RandomForestClassifier(n_estimators=40)
model.fit(X_train, y_train)
model.score(X_test,y_test)
```
